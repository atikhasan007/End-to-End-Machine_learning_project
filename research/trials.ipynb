```python
# ensure annotation demo

def get_product(x:int,y:int) -> int:

    return x * y
```

```python
get_product(x=5,y=6)
```

```python
get_product(x=2,y="3") # so it is worg 
```

```python
from ensure import ensure_annotations
```

```python
@ensure_annotations
def get_product(x:int, y:int) -> int:
    return x*y
```

```python
get_product(x=5,y=6)
```

```python
get_product(x=2,y="3") # so it is worg 
```
